# Visualization 1: Demo Data Marital Status by Gender and Ethnicity
- Compares the marital status of Women and Men on the Demo MIMIC III Data
- Gives ethnicity of different groups using Crosstabs
- Dynamic Bar Chart sizing based on number of individuals in each group
- Color coding across only ethnicity
- Wider spacing between pairs of similar marital statuses


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data from CSV files
patients = pd.read_csv('./mimic-iii-clinical-database-demo-1.4/PATIENTS.csv')
admissions = pd.read_csv('./mimic-iii-clinical-database-demo-1.4/ADMISSIONS.csv')

# Combine patients and admissions data
combined_df = pd.merge(patients, admissions, on='subject_id')

# Create separate dataframes for males and females
male_df = combined_df[combined_df['gender'] == 'M']
female_df = combined_df[combined_df['gender'] == 'F']

# Create crosstabs for each set
male_crosstab = pd.crosstab(male_df['marital_status'], male_df['ethnicity'])
female_crosstab = pd.crosstab(female_df['marital_status'], female_df['ethnicity'])

# Define color dictionary
color_dict = {
    'WHITE': 'pink',
    'ASIAN': 'yellow',
    'BLACK/AFRICAN AMERICAN': 'brown',
    'HISPANIC/LATINO - PUERTO RICAN': 'blue',
    'UNKNOWN/NOT SPECIFIED': 'grey',
    'AMERICAN INDIAN/ALASKA NATIVE FEDERALLY RECOGNIZED TRIBE': 'red',
    'UNABLE TO OBTAIN': 'black',
    'OTHER': 'green',
    'HISPANIC OR LATINO': 'blue'  # Added to match your data
}

# Get unique marital statuses
marital_statuses = sorted(set(male_crosstab.index).union(set(female_crosstab.index)))

# Create new x-axis labels alternating between F and M for each status
new_labels = []
bar_positions = []
for i, status in enumerate(marital_statuses):
    new_labels.extend([f'F - {status}', f'M - {status}'])
    bar_positions.extend([i*3, i*3 + 1])

# Create figure and axis
plt.figure(figsize=(15, 7))

# Get all unique ethnicities from both male and female data
all_ethnicities = sorted(set(male_crosstab.columns).union(set(female_crosstab.columns)))

# Create legend handles dictionary
legend_handles = {}

# Plot bars for each marital status
for status in marital_statuses:
    f_pos = bar_positions[new_labels.index(f'F - {status}')]
    m_pos = bar_positions[new_labels.index(f'M - {status}')]
    
    # Plot female data
    bottom = 0
    if status in female_crosstab.index:
        for ethnicity in female_crosstab.columns:
            value = female_crosstab.loc[status, ethnicity]
            if value > 0:
                bar = plt.bar(f_pos, value, bottom=bottom, 
                            color=color_dict.get(ethnicity, 'blue'))
                bottom += value
                # Store handle for legend
                if ethnicity not in legend_handles:
                    legend_handles[ethnicity] = bar

    # Plot male data
    bottom = 0
    if status in male_crosstab.index:
        for ethnicity in male_crosstab.columns:
            value = male_crosstab.loc[status, ethnicity]
            if value > 0:
                bar = plt.bar(m_pos, value, bottom=bottom,
                            color=color_dict.get(ethnicity, 'blue'))
                bottom += value
                # Store handle for legend
                if ethnicity not in legend_handles:
                    legend_handles[ethnicity] = bar

plt.title('Marital Status by Gender and Ethnicity')
plt.xlabel('Marital Status')
plt.ylabel('Count')

# Set x-axis labels
plt.xticks(bar_positions, new_labels, rotation=45, ha='right')

# Add the simplified legend
plt.legend(legend_handles.values(), legend_handles.keys(),
          loc='upper right', bbox_to_anchor=(1.15, 1))

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data from CSV files
patients = pd.read_csv('./mimic-iii-clinical-database-1.4/PATIENTS.csv.gz')
admissions = pd.read_csv('./mimic-iii-clinical-database-1.4/ADMISSIONS.csv.gz')

# Combine patients and admissions data
combined_df = pd.merge(patients, admissions, on='SUBJECT_ID')
combined_df.head()

# Create separate dataframes for males and females
male_df = combined_df[combined_df['GENDER'] == 'M']
female_df = combined_df[combined_df['GENDER'] == 'F']

# Create crosstabs for each set
male_crosstab = pd.crosstab(male_df['MARITAL_STATUS'], male_df['ETHNICITY'])
female_crosstab = pd.crosstab(female_df['MARITAL_STATUS'], female_df['ETHNICITY'])

# Define color dictionary with subclassifications having the same color
color_dict = {
    'WHITE': 'pink',
    'ASIAN': 'yellow',
    'BLACK/AFRICAN AMERICAN': 'brown',
    'HISPANIC/LATINO': 'orange',
    'UNKNOWN/NOT SPECIFIED': 'grey',
    'AMERICAN INDIAN/ALASKA NATIVE': 'red',
    'UNABLE TO OBTAIN': 'black',
    'OTHER': 'green'
}

# Function to map ethnicities to their base color
def map_ethnicity_color(ethnicity):
    for key in color_dict.keys():
        if key in ethnicity:
            return color_dict[key]
    return 'blue'  # Default color if no match found

# Get unique marital statuses
marital_statuses = sorted(set(male_crosstab.index).union(set(female_crosstab.index)))

# Create new x-axis labels alternating between F and M for each status
new_labels = []
bar_positions = []
for i, status in enumerate(marital_statuses):
    new_labels.extend([f'F - {status}', f'M - {status}'])
    bar_positions.extend([i*3, i*3 + 1])

# Create figure and axis
plt.figure(figsize=(15, 7))

# Get all unique ethnicities from both male and female data
all_ethnicities = sorted(set(male_crosstab.columns).union(set(female_crosstab.columns)))

# Create legend handles dictionary
legend_handles = {}

# Plot bars for each marital status
for status in marital_statuses:
    f_pos = bar_positions[new_labels.index(f'F - {status}')]
    m_pos = bar_positions[new_labels.index(f'M - {status}')]
    
    # Plot female data
    bottom = 0
    if status in female_crosstab.index:
        for ethnicity in female_crosstab.columns:
            value = female_crosstab.loc[status, ethnicity]
            if value > 0:
                color = map_ethnicity_color(ethnicity)
                bar = plt.bar(f_pos, value, bottom=bottom, color=color)
                bottom += value
                # Store handle for legend
                if ethnicity not in legend_handles:
                    legend_handles[ethnicity] = bar

    # Plot male data
    bottom = 0
    if status in male_crosstab.index:
        for ethnicity in male_crosstab.columns:
            value = male_crosstab.loc[status, ethnicity]
            if value > 0:
                color = map_ethnicity_color(ethnicity)
                bar = plt.bar(m_pos, value, bottom=bottom, color=color)
                bottom += value
                # Store handle for legend
                if ethnicity not in legend_handles:
                    legend_handles[ethnicity] = bar

plt.title('Marital Status by Gender and Ethnicity')
plt.xlabel('Marital Status')
plt.ylabel('Count')

# Set x-axis labels
plt.xticks(bar_positions, new_labels, rotation=45, ha='right')

# Add the legend at the bottom of the graph
plt.legend(legend_handles.values(), legend_handles.keys(),
           loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3)

plt.tight_layout()
plt.show()
```

# Visualization 2: Bubble Graph of Mortality Rate by Age Group
- Data is sourced from the restricted MIMIC III Dataset, rather than the Demo dataset
- **X-axis** is split up by the **age groups** 
- **Y-axis** is the calculated **mortality rate**
- Colors are determined by the calculated mortality rate, with low rates (near 0%) being **blue for low risk**, and high rates (nearing 18%) being marked **red for high risk**
- **Bubble Sizes** are determined by the **number of patients in the age group**, for population sizing to determine outliers

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors
from dateutil.relativedelta import relativedelta

# Load data from CSV files (adjust paths if needed)
patients = pd.read_csv('./mimic-iii-clinical-database-1.4/PATIENTS.csv.gz')
admissions = pd.read_csv('./mimic-iii-clinical-database-1.4/ADMISSIONS.csv.gz')

# Combine patients and admissions data
combined_df = pd.merge(patients, admissions, on='SUBJECT_ID')

# Convert ADMITTIME and DOB to datetime 
combined_df['ADMITTIME'] = pd.to_datetime(combined_df['ADMITTIME'], errors='coerce')
combined_df['DOB'] = pd.to_datetime(combined_df['DOB'], errors='coerce')

# Drop rows with NaT values resulting from parsing errors
combined_df.dropna(subset=['ADMITTIME', 'DOB'], inplace=True)

# Calculate age at admission (using relativedelta for accurate year calculation)
combined_df['ADMISSION_AGE'] = combined_df.apply(lambda row: relativedelta(row['ADMITTIME'], row['DOB']).years, axis=1)

# Filter out rows with unrealistic age values
combined_df = combined_df[(combined_df['ADMISSION_AGE'] >= 0) & (combined_df['ADMISSION_AGE'] <= 120)]

# --- Create age groups ---
bins = [0, 18, 30, 40, 50, 60, 70, 80, 90, 100]
labels = ['0-18', '19-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100']
combined_df['AGE_GROUP'] = pd.cut(combined_df['ADMISSION_AGE'], bins=bins, labels=labels, right=False)

# --- Calculate mortality rate for each age group ---
mortality_rate = combined_df.groupby('AGE_GROUP').agg(
    mortality_rate=('HOSPITAL_EXPIRE_FLAG', 'mean'),
    count=('HOSPITAL_EXPIRE_FLAG', 'count')
).reset_index()

mortality_rate.dropna(subset=['mortality_rate'], inplace=True)

# --- Convert age groups to numeric indices for plotting ---
age_group_numeric = np.arange(len(mortality_rate))

# --- Control Bubble Size ---
scaling_factor = .75  # Adjust this to change overall bubble size.

bubble_area = mortality_rate['count'] * scaling_factor
max_bubble_area = 20000  # Adjust this to set a maximum bubble area.
bubble_area = np.minimum(bubble_area, max_bubble_area)

# --- Plotting ---
plt.figure(figsize=(12, 8))

norm = mcolors.Normalize(vmin=mortality_rate['mortality_rate'].min(), 
                         vmax=mortality_rate['mortality_rate'].max())

# Use the calculated bubble areas in the scatter plot.
scatter = plt.scatter(age_group_numeric, mortality_rate['mortality_rate'], 
                      s=bubble_area,  # Now using bubble areas
                      c=mortality_rate['mortality_rate'], cmap="coolwarm", 
                      norm=norm,
                      alpha=0.7, edgecolors="black")

cbar = plt.colorbar(scatter)
cbar.set_label('Mortality Rate')
cbar.mappable.set_clim(vmin=mortality_rate['mortality_rate'].min(), vmax=mortality_rate['mortality_rate'].max())

plt.xticks(age_group_numeric, mortality_rate['AGE_GROUP'], rotation=45)
plt.title('Mortality Rate of a given case by Age Group (Bubble Size Represents Number of Patients)')
plt.xlabel('Age Group')
plt.ylabel('Mortality Rate')
plt.ylim(0, mortality_rate['mortality_rate'].max() * 1.1)
plt.grid(True, linestyle='--', alpha=0.5)

# Add labels for the bubble sizes (number of patients)
for i, row in mortality_rate.iterrows():
    plt.text(age_group_numeric[i], row['mortality_rate'], f"{row['count']}", 
             fontsize=9, ha='center', va='center', color='black')

plt.tight_layout()
plt.show()
```

# Visualization 3: Heatmap of Tests of Specimens and top 20 Organisms found
- Utilizes MIMIC III full dataset
- Compares tests for specimines and organisms located, to determine which tests can find which organisms
- Some tests are very generic and find a wide variety of Microorganisms, while others are more specific and only used to test for a few
    - Rows show test specificity
    - Columns show organism location preference
    - Darker squares show higher likelihood of finding a particular organism with a given test


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def create_microbio_heatmap(microbio_events_file_path, start_date=None, end_date=None, top_n=20):
    """Creates a heatmap of microbiology event data with custom color scheme and top N organisms."""

    try:
        microbio_events = pd.read_csv(microbio_events_file_path)
        microbio_events.columns = microbio_events.columns.str.lower()

        microbio_events['charttime'] = pd.to_datetime(microbio_events['charttime'], errors='coerce')
        microbio_events.dropna(subset=['charttime'], inplace=True)

        # --- Time Subset Selection (if applicable) ---
        if start_date:
            microbio_events = microbio_events[microbio_events['charttime'] >= start_date]
        if end_date:
            microbio_events = microbio_events[microbio_events['charttime'] <= end_date]

        # --- Identify Top N Organisms ---
        top_organisms = microbio_events['org_name'].value_counts().nlargest(top_n).index

        # --- Filter Data for Top Organisms ---
        microbio_events_filtered = microbio_events[microbio_events['org_name'].isin(top_organisms)]

        # --- Heatmap Creation ---
        heatmap_data = microbio_events_filtered.groupby(['spec_type_desc', 'org_name']).size().reset_index(name='count')
        pivot_table = heatmap_data.pivot_table(index='spec_type_desc', columns='org_name', values='count', fill_value=0)

        plt.figure(figsize=(15, 10))

        # --- Sequential Color Map (Corrected) ---
        cmap = sns.color_palette("light:#5F9EA0", as_cmap=True) # Light Blue to Red

        # Ensure 0 is mapped to white
        cmap.set_bad(color='white') # 0s will be white
        
        # Create a mask for 0 values (for proper display)
        mask = pivot_table == 0

        # Plot the heatmap with the custom color map and mask
        sns.heatmap(pivot_table, cmap=cmap, cbar=True, linewidths=.5, mask=mask, vmin=1)  # Set vmin=1 to exclude 0 from color scale

        plt.title(f'Heatmap of Specimen Types and Top {top_n} Organisms')
        plt.xlabel('Organism Name')
        plt.ylabel('Specimen Type Description')
        plt.tight_layout()

        return plt

    except FileNotFoundError:
        print(f"Error: File not found at {microbio_events_file_path}")
        return None
    except Exception as e:
        print(f"Error processing microbiology data: {str(e)}")
        return None


# Example Usage:
microbio_file = './mimic-iii-clinical-database-1.4/MICROBIOLOGYEVENTS.csv.gz'

try:
    heatmap_plt = create_microbio_heatmap(microbio_file, top_n=20)  # Show top 20 organisms
    if heatmap_plt:
        heatmap_plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import pandas as pd
import plotly.express as px

def create_microbio_plotly(microbio_events_file_path, start_date=None, end_date=None):
    """Creates interactive Plotly line charts with hover information."""

    try:
        microbio_events = pd.read_csv(microbio_events_file_path)
        microbio_events.columns = microbio_events.columns.str.lower()

        microbio_events['charttime'] = pd.to_datetime(microbio_events['charttime'], errors='coerce')
        microbio_events.dropna(subset=['charttime'], inplace=True)

        # --- Time Subset Selection ---
        if start_date:
            microbio_events = microbio_events[microbio_events['charttime'] >= start_date]
        if end_date:
            microbio_events = microbio_events[microbio_events['charttime'] <= end_date]

        microbio_events['year_month'] = microbio_events['charttime'].dt.to_period('M')
        agg_data = microbio_events.groupby(['year_month', 'spec_type_desc', 'org_name']).size().reset_index(name='count')
        agg_data['year_month'] = agg_data['year_month'].dt.strftime('%Y-%b')

        # --- Specimen Type Chart ---
        fig_spec = px.line(agg_data, x='year_month', y='count', color='spec_type_desc',
                           title='Frequency of Different Specimen Types Over Time')
        fig_spec.update_layout(xaxis_title='Year-Month', yaxis_title='Count',
                              legend_title='Specimen Type Description')

        # --- Organism Name Chart ---
        # Filter for organisms with max count > 10 (same as before)
        org_max_counts = agg_data.groupby('org_name')['count'].max()
        top_orgs = org_max_counts[org_max_counts > 10].index
        filtered_agg_data = agg_data[agg_data['org_name'].isin(top_orgs)]

        fig_org = px.line(filtered_agg_data, x='year_month', y='count', color='org_name',
                          title='Frequency of Different Organism Names Over Time (Max Count > 10)')
        fig_org.update_layout(xaxis_title='Year-Month', yaxis_title='Count',
                             legend_title='Organism Name')

        return fig_spec, fig_org  # Return both figures

    except FileNotFoundError:
        print(f"Error: File not found at {microbio_events_file_path}")
        return None, None
    except Exception as e:
        print(f"Error processing microbiology data: {str(e)}")
        return None, None


# Example Usage with Time Range Restriction (up to 40 years):
microbio_file = './mimic-iii-clinical-database-1.4/MICROBIOLOGYEVENTS.csv.gz'

try:
    # 1. Load the data and convert 'CHARTTIME' to datetime
    temp_df = pd.read_csv(microbio_file)
    temp_df['CHARTTIME'] = pd.to_datetime(temp_df['CHARTTIME'], errors='coerce')
    temp_df.dropna(subset=['CHARTTIME'], inplace=True) # Drop rows with invalid dates

    # 2. Sort data by 'CHARTTIME' to ensure correct order
    temp_df = temp_df.sort_values('CHARTTIME')

    # 3. Get the unique years and find the second year
    unique_years = temp_df['CHARTTIME'].dt.year.unique()

    if len(unique_years) >= 2: # Check if there are at least two years
        second_year = unique_years[1]

        # 4. Filter the DataFrame to include only the second year and onwards
        temp_df = temp_df[temp_df['CHARTTIME'].dt.year >= second_year]

        # 5. Find the earliest date in the filtered data
        earliest_date = temp_df['CHARTTIME'].min()

        # 6. Calculate the end date (3 years after the earliest date in the filtered data)
        end_date = earliest_date + pd.DateOffset(years=2)

        fig_spec, fig_org = create_microbio_plotly(microbio_file, start_date=earliest_date, end_date=end_date)
    if fig_spec and fig_org:
        fig_spec.show()
        fig_org.show()

except Exception as e:
    print(f"An error occurred: {e}")


except Exception as e:
    print(f"An error occurred: {e}")
```

# Visualization 4: ICD-9 Diagnosis Hierarchy

- Utilized a treemap to find hihg level categories and subcategories as codes
- Convert the numeric codes to their full descriptions
- Used plotly for Dynamic Plots which you can click into and out of

```python
import pandas as pd
import plotly.express as px
from collections import defaultdict

def create_diagnosis_treemap(diagnoses_file_path, d_icd_file_path):
    try:
        # Load the data
        diagnoses_df = pd.read_csv(diagnoses_file_path)
        d_icd_df = pd.read_csv(d_icd_file_path)
        
        # Ensure consistent column names
        diagnoses_df.columns = diagnoses_df.columns.str.lower()
        d_icd_df.columns = d_icd_df.columns.str.lower()
        
        # Create mappings for both full codes and major categories
        def get_major_code(code):
            if pd.isna(code) or not isinstance(code, str):
                return None
            code = str(code).strip()
            if code.startswith('E'):
                return code[:4]  # E-codes are handled differently
            if code.startswith('V'):
                return code[:3]  # V-codes are handled differently
            return code[:3].zfill(3)  # Regular codes
        
        # Add major code column to d_icd_df
        d_icd_df['major_code'] = d_icd_df['icd9_code'].apply(get_major_code)
        
        # Create separate mappings for full codes and major categories
        full_code_descriptions = dict(zip(d_icd_df['icd9_code'], d_icd_df['short_title']))
        
        # Create major category descriptions by taking the first occurrence of each major code
        major_descriptions = d_icd_df.drop_duplicates('major_code').set_index('major_code')['short_title'].to_dict()
        
        # Merge diagnoses with ICD descriptions
        df = diagnoses_df.merge(d_icd_df, on='icd9_code', how='left')
        
        def get_icd9_hierarchy(row):
            code = row['icd9_code']
            description = row['short_title']
            
            if pd.isna(code) or not isinstance(code, str):
                return ['Unknown']
            
            code = code.strip()
            major_code = get_major_code(code)
            
            if code.startswith('E'):
                major_desc = major_descriptions.get(major_code, 'External Injury')
                return ['External Causes', f"{code}: {description}"]
            if code.startswith('V'):
                major_desc = major_descriptions.get(major_code, 'Supplementary Factors')
                return ['Supplementary Classification', f"{code}: {description}"]
            
            # Get major category description
            major_desc = major_descriptions.get(major_code, 'Unknown Category')
            
            # Return hierarchy with descriptions
            if len(code) > 3:
                return [f"{major_code}: {major_desc}", f"{code}: {description}"]
            else:
                return [f"{code}: {description}"]

        # Create hierarchy columns
        df['hierarchy'] = df.apply(get_icd9_hierarchy, axis=1)
        
        # Count diagnoses at each level
        hierarchy_counts = defaultdict(int)
        for _, row in df.iterrows():
            for level in row['hierarchy']:
                hierarchy_counts[level] += 1
        
        # Create dataframe for treemap
        treemap_data = []
        for label, count in hierarchy_counts.items():
            if ':' in label:
                code = label.split(':')[0].strip()
                if len(code) <= 3 and not code.startswith(('E', 'V')):  # Major category
                    treemap_data.append({
                        'code': label,
                        'parent': '',
                        'value': count
                    })
                else:  # Subcategory
                    major_code = get_major_code(code)
                    major_desc = major_descriptions.get(major_code, 'Unknown')
                    parent = f"{major_code}: {major_desc}"
                    treemap_data.append({
                        'code': label,
                        'parent': parent,
                        'value': count
                    })
            else:  # Special categories (External Causes, Supplementary)
                if label in ['External Causes', 'Supplementary Classification']:
                    treemap_data.append({
                        'code': label,
                        'parent': '',
                        'value': count
                    })
                else:
                    treemap_data.append({
                        'code': label,
                        'parent': label.split(' ')[0],
                        'value': count
                    })
        
        treemap_df = pd.DataFrame(treemap_data)
        
        # Create treemap visualization
        fig = px.treemap(
            treemap_df,
            path=['parent', 'code'],
            values='value',
            title='ICD-9 Diagnosis Hierarchy Distribution in MIMIC III',
            width=1200,
            height=900
        )
        
        # Update layout
        fig.update_traces(
            textinfo='label+value',
            hovertemplate='<b>%{label}</b><br>Count: %{value}<br>Percentage: %{percentParent:.1%}<extra></extra>'
        )
        
        fig.update_layout(
            margin=dict(t=50, l=25, r=25, b=25)
        )
        
        return fig
    
    except Exception as e:
        print(f"Error processing data: {str(e)}")
        raise

# Example usage:
fig = create_diagnosis_treemap('./mimic-iii-clinical-database-1.4/DIAGNOSES_ICD.csv.gz', './mimic-iii-clinical-database-1.4/D_ICD_DIAGNOSES.csv.gz')
fig.show()
```

# Visualization 5: Bonus using pygal to export Vectors

- Used Pygal which was not covered in lectures
- Visualized a basic bar chart, but with two distinct groups, people who were transfered and the general population

```python
import pandas as pd
import pygal

def visualize_patient_conditions_pygal(icustays_file, diagnoses_file, d_icd_file, top_n_conditions=20):
    """Visualizes top N patient conditions using Pygal (normalized frequencies)."""

    try:
        icustays_df = pd.read_csv(icustays_file)
        diagnoses_df = pd.read_csv(diagnoses_file)
        diagnoses_labels = pd.read_csv(d_icd_file)

        # Convert column names to lowercase for consistency (recommended)
        icustays_df.columns = icustays_df.columns.str.lower()
        diagnoses_df.columns = diagnoses_df.columns.str.lower()
        diagnoses_labels.columns = diagnoses_labels.columns.str.lower()

        icd_dict = diagnoses_labels.drop_duplicates(subset='icd9_code').set_index('icd9_code')['short_title'].to_dict()

        merged_df = pd.merge(icustays_df, diagnoses_df, on=['subject_id', 'hadm_id'])
        merged_df['diagnosis_label'] = merged_df['icd9_code'].map(icd_dict)

        # Calculate and normalize frequencies (for both transferred and all patients)
        condition_frequency_transferred = (
            merged_df[merged_df['first_careunit'] != merged_df['last_careunit']]
            .groupby('diagnosis_label')
            .size()
            .reset_index(name='frequency')
        )
        condition_frequency_all = (
            merged_df.groupby('diagnosis_label')
            .size()
            .reset_index(name='frequency')
        )

        n_transferred = len(merged_df[merged_df['first_careunit'] != merged_df['last_careunit']]['subject_id'].unique())
        n_all = len(merged_df['subject_id'].unique())

        condition_frequency_transferred['frequency_normalized'] = condition_frequency_transferred['frequency'] / n_transferred
        condition_frequency_all['frequency_normalized'] = condition_frequency_all['frequency'] / n_all

        # Select top N conditions (considering both groups)
        top_conditions_transferred = condition_frequency_transferred.nlargest(top_n_conditions, 'frequency')['diagnosis_label'].tolist()
        top_conditions_all = condition_frequency_all.nlargest(top_n_conditions, 'frequency')['diagnosis_label'].tolist()
        top_conditions = list(set(top_conditions_transferred + top_conditions_all)) # Combine and remove duplicates
        merged_df = merged_df[merged_df['diagnosis_label'].isin(top_conditions)] # Filter the merged df

        # Recalculate normalized frequencies based on the filtered data
        condition_frequency_transferred = (
            merged_df[merged_df['first_careunit'] != merged_df['last_careunit']]
            .groupby('diagnosis_label')
            .size()
            .reset_index(name='frequency')
        )
        condition_frequency_all = (
            merged_df.groupby('diagnosis_label')
            .size()
            .reset_index(name='frequency')
        )
        condition_frequency_transferred['frequency_normalized'] = condition_frequency_transferred['frequency'] / n_transferred
        condition_frequency_all['frequency_normalized'] = condition_frequency_all['frequency'] / n_all

        condition_frequency = pd.merge(
            condition_frequency_transferred, condition_frequency_all, on='diagnosis_label', how='outer'
        ).fillna(0)

        # Create the bar chart
        bar_chart = pygal.Bar(title=f"Normalized Frequency of Top {top_n_conditions} Conditions (Transferred vs. All Patients)")

        for condition in top_conditions:  # Iterate through the top conditions
            freq_transferred = condition_frequency.loc[condition_frequency['diagnosis_label'] == condition, 'frequency_normalized_x'].iloc[0] if (condition_frequency['diagnosis_label'] == condition).any() else 0
            freq_all = condition_frequency.loc[condition_frequency['diagnosis_label'] == condition, 'frequency_normalized_y'].iloc[0] if (condition_frequency['diagnosis_label'] == condition).any() else 0


            bar_chart.add(f"{condition} (Transferred)", freq_transferred)
            bar_chart.add(f"{condition} (All)", freq_all)

        bar_chart.render_to_file("condition_frequencies.svg")  # Save as SVG

        print(f"Visualization saved to condition_frequencies.svg")

    except FileNotFoundError:
        print("Error: One or more files not found. Please check the file paths.")
    except pd.errors.EmptyDataError:
        print("Error: One or more of the CSV files are empty.")
    except KeyError as e:
        print(f"Error: A key error occurred.  Check if the necessary columns ({e}) are present in your dataframes.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


# Example Usage:
icustays_file = './mimic-iii-clinical-database-1.4/ICUSTAYS.csv.gz'
diagnoses_file = './mimic-iii-clinical-database-1.4/DIAGNOSES_ICD.csv.gz'
d_icd_file = './mimic-iii-clinical-database-1.4/D_ICD_DIAGNOSES.csv.gz'

visualize_patient_conditions_pygal(icustays_file, diagnoses_file, d_icd_file)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches

def visualize_patient_conditions(icustays_file, diagnoses_file, d_icd_file, transferred_only=True, min_frequency=1, top_n_conditions=None):
    """Visualizes patient conditions with filtering for top conditions."""

    try:
        icustays_df = pd.read_csv(icustays_file)
        diagnoses_df = pd.read_csv(diagnoses_file)
        diagnoses_labels = pd.read_csv(d_icd_file)

        icustays_df.columns = icustays_df.columns.str.lower()
        diagnoses_df.columns = diagnoses_df.columns.str.lower()
        diagnoses_labels.columns = diagnoses_labels.columns.str.lower()

        icd_dict = diagnoses_labels.drop_duplicates(subset='icd9_code').set_index('icd9_code')['short_title'].to_dict()

        if transferred_only:
            icustays_df = icustays_df[icustays_df['first_careunit'] != icustays_df['last_careunit']]

        merged_df = pd.merge(icustays_df, diagnoses_df, on=['subject_id', 'hadm_id'])
        merged_df['diagnosis_label'] = merged_df['icd9_code'].map(icd_dict)

        if top_n_conditions is not None:
            condition_frequency = merged_df.groupby('diagnosis_label').size().reset_index(name='frequency')
            top_conditions = condition_frequency.nlargest(top_n_conditions, 'frequency')['diagnosis_label'].tolist()
            merged_df = merged_df[merged_df['diagnosis_label'].isin(top_conditions)]

        pivot_table = merged_df.pivot_table(index='subject_id', columns='diagnosis_label', aggfunc='size', fill_value=0)
        pivot_table = pivot_table.applymap(lambda x: 1 if x > 0 else 0)

        plt.figure(figsize=(14, 21))

        plt.subplot(3, 1, 1)
        sns.heatmap(pivot_table, cmap='RdYlGn_r', cbar=False)
        plt.title('Heatmap of Conditions for {} Patients (Top {} Conditions)'.format(
            "Transferred" if transferred_only else "All",
            top_n_conditions if top_n_conditions is not None else "All"
        ))
        plt.xlabel('Diagnosis Label')
        plt.ylabel('Patient ID')
        plt.xticks(rotation=90)

        red_patch = mpatches.Patch(color='red', label='Has Condition')
        green_patch = mpatches.Patch(color='green', label='No Condition')
        plt.legend(handles=[red_patch, green_patch], loc='upper right', bbox_to_anchor=(1.15, 1))

        plt.subplot(3, 1, 2)
        condition_frequency = pivot_table.sum().reset_index()
        condition_frequency.columns = ['diagnosis_label', 'frequency']
        sns.lineplot(data=condition_frequency, x='diagnosis_label', y='frequency', marker='o')
        plt.title('Frequency of Conditions in {} Patients (Top {} Conditions)'.format(
            "Transferred" if transferred_only else "All",
            top_n_conditions if top_n_conditions is not None else "All"
        ))
        plt.xlabel('Diagnosis Label')
        plt.ylabel('Frequency')
        plt.xticks(rotation=90)

        # Frequent conditions heatmap
        plt.subplot(3, 1, 3)
        frequent_conditions = condition_frequency[condition_frequency['frequency'] > min_frequency]
        sns.heatmap(frequent_conditions[['frequency']].T, cmap='Reds', cbar=True, annot=True, fmt='d')
        plt.title('Heatmap of Frequent Conditions (Frequency > {})'.format(min_frequency))
        plt.xlabel('Diagnosis Label')
        plt.yticks([])

        # Get the number of conditions (columns in the heatmap)
        num_conditions = len(frequent_conditions)

        # Set x-ticks to be at the center of each cell
        tick_positions = np.arange(0.5, num_conditions + 0.5, 1)

        # Set x-tick labels and rotate them
        plt.xticks(ticks=tick_positions, labels=frequent_conditions['diagnosis_label'], rotation=90, ha='center')

        plt.tight_layout()
        plt.show()

    except FileNotFoundError:
        print("Error: One or more files not found.")
    except Exception as e:
        print(f"An error occurred: {e}")


# Example Usage:
icustays_file = './mimic-iii-clinical-database-1.4/ICUSTAYS.csv.gz'
diagnoses_file = './mimic-iii-clinical-database-1.4/DIAGNOSES_ICD.csv.gz'
d_icd_file = './mimic-iii-clinical-database-1.4/D_ICD_DIAGNOSES.csv.gz'

# For transferred patients only, top 20 conditions:
visualize_patient_conditions(icustays_file, diagnoses_file, d_icd_file, top_n_conditions=20)

# For all patients, top 20 conditions, min frequency 2:
visualize_patient_conditions(icustays_file, diagnoses_file, d_icd_file, transferred_only=False, top_n_conditions=20, min_frequency=2)
```
